```python
import json
import re
```

```python
JSON_data = []
```

```python
def loadConllFile(filename):
    with open (filename, "r") as file:
        return(file.readlines())
```

```python
JSON_data = loadConllFile("json/ensar_son.jsonl")
```

```python
datass = []
for data in JSON_data:
    loads = json.loads(data)
    text = loads["text"] 
    
    label_texts = []
    label_ents = []

    for l in loads["label"]:
        label_texts.append(text[l[0]:l[1]])
        label_ents.append(l[2])
    
    texts  = text.split(".")

    for t in texts:
        t = t.lstrip()
        startpos = 0
        epos = 0
        spos =0
        tags = []
        for ind, lt in enumerate(label_texts):
            spos =t.find(lt, startpos) 

            if spos >= 0:
                epos = spos+len(lt)
                startpos = epos
                tags.append([spos, epos, label_ents[ind]])
        
        if len(tags) > 0: 
            datass.append({"text": t, "label": tags})

print(len(datass))


for d in datass:
    print(json.dumps(d,ensure_ascii=False)) 

    
```

```python
with open ("json/ensar.json", "w+") as nerfile:
    for n in datass:
        nerfile.writelines(json.dumps(n, ensure_ascii= False)+"\n")
```
